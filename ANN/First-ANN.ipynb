```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import csv


class ReadCorpus(object):
    def __init__(self,csvfile,dumm_columns=[]):
        csv.field_size_limit(10**9)
        self.reader = csv.reader(open(csvfile,"r"),delimiter=",", quoting=csv.QUOTE_MINIMAL)
        self.header = self.reader.__next__()
        apaga = csv.reader(open(csvfile,"r"),delimiter=",", quoting=csv.QUOTE_MINIMAL)
        apaga.__next__()
        self.totalsents = (len(list(apaga)))
    
    def __iter__(self):
        index = 0
        for index,row in enumerate(self.reader):
            dic = dict(zip(self.header, row))
            gender=pd.get_dummies(dic['Gender'],columns=['Male','Female'])
            print(gender)
            return
            print("Progress:", (index+1), "/", self.totalsents)
            index += 1
            yield  dic #['data']
```

```python
# Importing the dataset
read = ReadCorpus('Churn_Modelling.csv')
for r in read:
    print(r)
```

```python
dataset = pd.read_csv('Churn_Modelling.csv',delimiter=",")


# Encoding categorical data
geo=pd.get_dummies(dataset['Geography'],drop_first="true")
combine = pd.concat([geo, dataset], axis=1)
gender=pd.get_dummies(combine['Gender'])
combine = pd.concat([gender, combine], axis=1)
combine = combine.drop(columns=['Geography','Gender','RowNumber','CustomerId','Surname'])

X = combine.iloc[:,0:11].values
print(len(X[1]))
y = combine.iloc[:,-1].values
print(y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python

# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
